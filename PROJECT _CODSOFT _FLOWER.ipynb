```python
import warnings
import joblib
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
iris=load_iris()
X=pd.DataFrame(iris.data,columns=iris.feature_names)
y=pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=1)
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig=plt.figure(figsize=(25,20))
_=tree.plot_tree(dt,feature_names=iris.feature_names,
                 class_names=iris.target_names,
                 filled=True)
```

```python
y_pred=dt.predict(X_test)
```

```python
species=np.array(y_test).argmax()
predictions=np.array(y_pred).argmax()
species = y_test
predictions = y_pred

confusion_matrix(species,predictions)
```

Jaccard Index

```python
from sklearn.metrics import classification_report

# Assuming y_test and y_pred contain the true and predicted labels
report = classification_report(y_test, y_pred)
print(report)
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
